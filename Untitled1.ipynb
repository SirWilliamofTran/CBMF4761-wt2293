Generate Commercial Mab Distributions

```python
import pandas as pd

data = pd.read_csv('SeqStruc1.csv',header=0,converters={'Heavy Sequence':str,'Light Sequence':str})
df = pd.DataFrame(data, columns= ['Heavy Sequence','Light Sequence'])
Heavy = df['Heavy Sequence']
Light = df['Light Sequence']
data = pd.read_csv('motifs.csv',header=0,converters={'Motif':str,'Risk/Impact':str})
df = pd.DataFrame(data, columns= ['Region','Motif','Risk/Impact'])
Motif = df['Motif']
Risk = df['Risk/Impact']
Region = df['Region']
data = pd.read_csv('antibody.csv',header=0,converters={'VDJ':str,'CDR3':str})
df = pd.DataFrame(data, columns= ['VDJ','CDR3'])
VDJ = df['VDJ']
```

```python
import re 
import numpy as np

CMabs = np.zeros((len(Heavy),6))
Oxidation = 0
Isomerization = 0
Glycosylation = 0
Disulfide = 0
Deamidation = 0
Aggregation = 0
Read = [Oxidation, Isomerization, Glycosylation, Disulfide, Deamidation, Aggregation]

for j in range(0,len(Heavy)):
    for i in range(0,len(Motif)):
        if re.findall(Motif[i],Heavy[j]):
            if 0 <= i <= 1:
                Oxidation+=len(re.findall(Motif[i],Heavy[j]))
            elif 2 <= i <= 4:
                Isomerization+=len(re.findall(Motif[i],Heavy[j]))
            elif 5 <= i <= 7:
                Glycosylation+=len(re.findall(Motif[i],Heavy[j]))
            elif i == 8:
                Disulfide+=len(re.findall(Motif[i],Heavy[j])) 
            elif 9 <= i <= 17:
                Deamidation+=len(re.findall(Motif[i],Heavy[j]))
            else:
                Aggregation+=len(re.findall(Motif[i],Heavy[j]))     
        if re.findall(Motif[i],Light[j]):
            if 0 <= i <= 1:
                Oxidation+=len(re.findall(Motif[i],Heavy[j]))
            elif 2 <= i <= 4:
                Isomerization+=len(re.findall(Motif[i],Heavy[j]))
            elif 5 <= i <= 7:
                Glycosylation+=len(re.findall(Motif[i],Heavy[j]))
            elif i == 8:
                Disulfide+=len(re.findall(Motif[i],Heavy[j])) 
            elif 9 <= i <= 17:
                Deamidation+=len(re.findall(Motif[i],Heavy[j]))
            else:
                Aggregation+=len(re.findall(Motif[i],Heavy[j]))
    
    Read = [Oxidation/2, Isomerization/2, Glycosylation/2, Disulfide/2, Deamidation/2, Aggregation/2]         
    CMabs[j] = Read         
    Oxidation = 0
    Isomerization = 0
    Glycosylation = 0
    Disulfide = 0
    Deamidation = 0
    Aggregation = 0
    
```

```python
import matplotlib.pyplot as plt

Q5a = np.quantile(CMabs[:,0],0.05)
Q95a = np.quantile(CMabs[:,0],0.95)
Q5b = np.quantile(CMabs[:,1],0.05)
Q95b = np.quantile(CMabs[:,1],0.95)
Q5c = np.quantile(CMabs[:,2],0.05)
Q95c = np.quantile(CMabs[:,2],0.95)
Q5d = np.quantile(CMabs[:,3],0.05)
Q95d = np.quantile(CMabs[:,3],0.95)
Q5e = np.quantile(CMabs[:,4],0.05)
Q95e = np.quantile(CMabs[:,4],0.95)
Q5f = np.quantile(CMabs[:,5],0.05)
Q95f = np.quantile(CMabs[:,5],0.95)

plt.subplot(1,2,1)
plt.hist(CMabs[:,0])
plt.title('Possible Oxidated Residues')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5a, ls='--', c='y')
plt.axvline(Q95a, ls='--', c='y')

plt.subplot(1,2,2)
plt.hist(CMabs[:,1])
plt.title('Possible Isomerized Residues')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5b, ls='--', c='y')
plt.axvline(Q95b, ls='--', c='y')

plt.figure()
plt.subplot(1,2,1)
plt.hist(CMabs[:,2])
plt.title('Possible Glycosylated Residues')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5c, ls='--', c='y')
plt.axvline(Q95c, ls='--', c='y')

plt.subplot(1,2,2)
plt.hist(CMabs[:,4])
plt.title('Possible Deamidation Sites')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5e, ls='--', c='y')
plt.axvline(Q95e, ls='--', c='y')
plt.figure()
plt.subplot(1,2,1)
plt.hist(CMabs[:,3])
plt.title('Number of Cysteines')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5d, ls='--', c='y')
plt.axvline(Q95d, ls='--', c='y')
plt.subplot(1,2,2)
plt.hist(CMabs[:,5])
plt.title('Aggregation Prone Regions')
plt.ylabel('Number of Commercialized Mabs')
plt.xlabel('Number of Motifs')
plt.axvline(Q5f, ls='--', c='y')
plt.axvline(Q95f, ls='--', c='y')
```
